```python
import torch
import numpy as np
```

```python
data = torch.randn(1,28,28)
```

```python
layer = torch.nn.Conv2d(in_channels=1, out_channels=10, kernel_size=3, stride=1)
layer
```

```python
layer.weight
```

```python
layer.weight.detach()
#  Autograd 추적을 제외시키고 가중치 값을 고정하거나 복사하는 데 사용
```

```python
layer.weight.shape
```

```python
layer(data).shape, layer(data)
```

```python
layer.bias, layer.bias.detach()
```

```python
data.unsqueeze(0).shape
```

```python
out = layer(data.unsqueeze(0))
out.shape
```

```python
flatten_data = data.view(28*28)
flatten_data.shape
```

```python
result = torch.nn.Linear(784, 5)(flatten_data)
result
```

```python
plt.plot(result.detach());
```

```python
softmax = F.softmax(result)
softmax
```

```python
plt.plot(softmax)
```

```python
plt.plot(softmax.detach());
```
