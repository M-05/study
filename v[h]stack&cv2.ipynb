```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((512,512,3),dtype=np.int32)

text = 'A B C D E F G'
text2 = 'H I J K L M N'
text3 = 'O P Q R S T U'
text4 = 'V W X Y Z'
text5= 'a b c d e f g'
text6 = 'h i j k l m n'
text_array = [text,text2,text3,text4,text5,text6]
text7='A'
text8='S'
text9='b'
text10='f'
```

```python
count = 1
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
font
```

```python
for t in text_array:
    if count>4:
        font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
    org = (50, count * 80)
    cv2.putText(img=img, text=t, org=org,
                fontFace=font, fontScale= 1.7,
                color=(255,255,255), thickness=7)
    count = count+1
```
